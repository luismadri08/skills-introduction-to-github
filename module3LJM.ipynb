Exercise 1: Salary Tax Calculator

Calculate 2026 Federal tax based on income brackets:
35% for incomes over $256,225 ($512,450 for married couples filing jointly);
32% for incomes over $201,775 ($403,550 for married couples filing jointly);
24% for incomes over $105,700 ($211,400 for married couples filing jointly);
22% for incomes over $50,400 ($100,800 for married couples filing jointly);
12% for incomes over $12,400 ($24,800 for married couples filing jointly).

```python
to_pay = 0

def federal_tax():
    marital_status = int(input('If you are married, enter 2 and if you are single, enter 1: '))

    if marital_status == 1:
        print('We will proceed with your income and calculate taxes as single.')
        income = int(input('What was your annual income? '))

        if income > 256225:
            to_pay = income * 0.35
            print('Based on your income you have to pay 35% and it is equivalent to =', to_pay)
        elif income > 201775:
            to_pay = income * 0.32
            print('Based on your income you have to pay 32% and it is equivalent to =', to_pay)
        elif income > 105700:
            to_pay = income * 0.24
            print('Based on your income you have to pay 24% and it is equivalent to =', to_pay)
        elif income > 50400:
            to_pay = income * 0.22
            print('Based on your income you have to pay 22% and it is equivalent to =', to_pay)
        elif income > 12400:
            to_pay = income * 0.12
            print('Based on your income you have to pay 12% and it is equivalent to =', to_pay)
        else:
            print('Input not valid')

    elif marital_status == 2:
        print('We will proceed with your income and calculate taxes as married.')
        income = int(input('What was your annual income? '))

        if income > 512450:
            to_pay = income * 0.35
            print('Based on your income you have to pay 35% and it is equivalent to =', to_pay)
        elif income > 403550:
            to_pay = income * 0.32
            print('Based on your income you have to pay 32% and it is equivalent to =', to_pay)
        elif income > 211400:
            to_pay = income * 0.24
            print('Based on your income you have to pay 24% and it is equivalent to =', to_pay)
        elif income > 100800:
            to_pay = income * 0.22
            print('Based on your income you have to pay 22% and it is equivalent to =', to_pay)
        elif income > 24800:
            to_pay = income * 0.12
            print('Based on your income you have to pay 12% and it is equivalent to =', to_pay)
        else:
            print('Input not valid')

    else:
        print('Invalid marital status entered. Please enter 1 or 2.')
        return None

    return to_pay

federal_tax()




```
